```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
columns = ['Pregnancies', 'Glucose', 'BloodPressure','SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction','Age', 'Outcome']
df = pd.read_csv(url, names=columns)
```

#### Display the first few rows

```python

df.head()
```

#### Descriptive statistics

```python

df.describe()
```

#### Visualising the distribution of features

```python

plt.figure(figsize=(10,8))
for i,col in enumerate(df.columns[:-1]) :
    plt.subplot(3,3,i+1)
    sns.histplot(df[col],kde = True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

#### Correlation heatmap

```python
plt.Figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#### Analyzing patient demographics (Age and Pregnancies)

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Distribution of Age')
plt.subplot(1, 2, 2)
sns.histplot(df['Pregnancies'], kde=True, bins=10)
plt.title('Distribution of Pregnancies')
plt.tight_layout()
plt.show()
```

#### Analyzing health indicators (Glucose, Blood Pressure, BMI)

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(df['Glucose'], kde=True)
plt.title('Distribution of Glucose')
plt.subplot(1, 3, 2)
sns.histplot(df['BloodPressure'], kde=True)
plt.title('Distribution of Blood Pressure')
plt.subplot(1, 3, 3)
sns.histplot(df['BMI'], kde=True)
plt.title('Distribution of BMI')
plt.tight_layout()
plt.show()
```

#### Scatter plot matrix to identify trends

```python
sns.pairplot(df, diag_kind='kde', markers='+', plot_kws={'alpha':0.3})
plt.suptitle('Scatter Plot Matrix', y=1.02)
plt.show()
```

#### Analyzing the relationship between Glucose and Outcome

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title('Glucose Levels by Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('Glucose Level')
plt.show()
```

#### Analyzing the relationship between BMI and Outcome

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title('BMI Levels by Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('BMI Level')
plt.show()
```
